# 04 Directions 5-7

05 The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

On average, loyal customers tend to buy products at a higher price than new and regular customer. However, the difference is very minor

06 The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Creating a spendings column
df_ords_prods_merge['spendings'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform (np.max)
```

```python
df_ords_prods_merge.head()
```

```python
# Deriving columns with loc()
df_ords_prods_merge.loc[df_ords_prods_merge['spendings'] < 10, 'spending_flag'] = 'Low spender'
df_ords_prods_merge.loc[df_ords_prods_merge['spendings'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge[['user_id', 'spending_flag', 'spendings']].head(100)
```

07 Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Creating a spendings column
df_ords_prods_merge['order frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform (np.max)
```

```python
df_ords_prods_merge.head()
```

```python
# Deriving columns with loc()

df_ords_prods_merge.loc[df_ords_prods_merge['order frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[df_ords_prods_merge['order frequency'] > 10, 'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['order frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_ords_prods_merge[['user_id', 'order_frequency_flag', 'order frequency']].head(100)
```
